***테스트 과정***

테스트할 원본 이미지들을 test_original 폴더에 저장

-> 테스트 이미지 파일명을 번호.jpg 형식으로 변경

-> 테스트 원본 이미지를 256x256으로 resize해서 test/check 폴더에 저장 (+ 원본 크기 저장해둠)

-> 테스트 진행

-> 원본 사이즈로 resize


```
import sys 
import os
import cv2
import numpy as np
```

```
# 테스트할 이미지는 test_path 경로에 저장해야함
# 테스트 원본 이미지 -> test_original

test_path = '/content/drive/MyDrive/StarGAN/data/custom/test_original/'
files = os.listdir(test_path) # 현재 위치의 파일 목록  
#print(files)
file_num = len(files)
print("테스트 파일 수 : %d" % file_num)
```

```
# # test 이미지 파일명을 번호.jpg 형식으로 변경 (이미 되어있어 주석처리)

# count = 1
# for name in files:
#   new_name = f'{count}.jpg'
#   os.rename(test_path+name, test_path+new_name)
#   count += 1

# files = os.listdir(test_path) # 현재 위치의 파일 목록  
# print(sorted(files))
```

```
# 원본 이미지를 256x256으로 resize

org_shape = []
#for i, f in enumerate(files):
for i in range(1, file_num+1):
  try:
    path = test_path + str(i)+'.jpg'
    
    img = cv2.imread(path, cv2.IMREAD_COLOR)

    org_shape.append( (img.shape[0], img.shape[1]) )  # 원본 크기 저장(튜플)

    img_resize = cv2.resize(img, dsize=(256, 256), interpolation=cv2.INTER_CUBIC)  

    write_path = '/content/drive/MyDrive/StarGAN/data/custom/test/check/' + str(i)+'.jpg'
    cv2.imwrite(write_path, img_resize)

  except Exception as e:  # resize 불가
      print('Error: ' + str(i) + '.jpg ' + str(e))
```
